```python
import numpy as np
import matplotlib.pyplot as plt

# path to saved mel-spectrogram .npy file
spec_path = r"C:\MTP\new_model\features\spectrogram\atlantic_spotted_dolphin\6102500B_mel.npy"

# load spectrogram
spec = np.load(spec_path)

# if shape is (1, 128, T), remove channel dimension
if spec.ndim == 3:
    spec = spec.squeeze(0)

print("Spectrogram shape:", spec.shape)

plt.figure(figsize=(12, 5))
plt.imshow(spec, aspect="auto", origin="lower", cmap="magma")
plt.colorbar(label="Amplitude / Log-mel energy")
plt.xlabel("Time frames")
plt.ylabel("Mel frequency bins")
plt.title("Mel-spectrogram")
plt.tight_layout()
plt.show()
```
