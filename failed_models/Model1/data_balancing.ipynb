```python
import numpy as np
import json
import os
```

```python
def augment_spectrogram(spec):
    aug_spec = spec.copy()
    chnl, rows, cols = spec.shape
    
    # 1. Time Masking (20%)
    t_width = int(0.2 * cols)
    t0 = np.random.randint(0, cols - t_width)
    aug_spec[:, t0:t0+t_width] = 0
    
    # 2. Frequency Masking (20%)
    f_width = int(0.2 * rows)
    f0 = np.random.randint(0, rows - f_width)
    aug_spec[f0:f0+f_width, :] = 0
    
    # 3. Noise Addition (10% of max intensity)
    noise = np.random.normal(0,0.05*np.std(spec), spec.shape)
    aug_spec = aug_spec + noise
    
    return aug_spec
```

```python
targ_len=len(os.listdir("features/spectrogram/killer_whale"))

dic="features/spectrogram"
for sp in os.listdir(dic):
    folder=os.path.join(dic,sp)
    curr_len=len(os.listdir(folder))
    for fl in os.listdir(folder):
        sp=np.load(os.path.join(folder,fl))        
        for i in range(targ_len//curr_len):
            new_sp=augment_spectrogram(sp)

            nam=fl.split("_")[0]+f"augmented{i}_mel"
            np.save(os.path.join(folder,nam),new_sp)
```
