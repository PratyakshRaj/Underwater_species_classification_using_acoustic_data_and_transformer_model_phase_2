```python
plt.scatter([i for i in range(1,26)],train_f1,c="blue",label="train_f1")
plt.scatter([i for i in range(1,26)],test_f1,c="red",label="test_f1")
plt.ylabel("f1")
plt.xlabel("epoch")
plt.legend()
```

```python
plt.scatter([i for i in range(1,26)],train_acc,c="blue",label="train_acc")
plt.scatter([i for i in range(1,26)],test_acc,c="red",label="test_acc")
plt.ylabel("acc")
plt.xlabel("epoch")
plt.legend()
```

results all folds

```python
train_loss=[]
train_acc=[]
train_f1=[]
test_loss=[]
test_acc=[]
test_f1=[]
for i in range(1,6):
    train_loss.append(spec_result[f"fold_{i}"]["epoch_25"]["train_loss"])
    train_acc.append(spec_result[f"fold_{i}"][f"epoch_25"]["train_accuracy"])
    train_f1.append(spec_result[f"fold_{i}"][f"epoch_25"]["train_f1"])
    test_loss.append(spec_result[f"fold_{i}"][f"epoch_25"]["val_loss"])
    test_acc.append(spec_result[f"fold_{i}"][f"epoch_25"]["val_accuracy"])
    test_f1.append(spec_result[f"fold_{i}"][f"epoch_25"]["val_f1"])
```

```python
plt.scatter([i for i in range(1,6)],train_loss,c="blue",label="train_loss")
plt.scatter([i for i in range(1,6)],test_loss,c="red",label="test_loss")
plt.ylabel("loss")
plt.xlabel("fold")
plt.ylim(0,1)
plt.legend()
print("average train loss:",np.average(train_loss))
print("average test loss:",np.average(test_loss))
```

```python
plt.scatter([i for i in range(1,6)],train_f1,c="blue",label="train_f1")
plt.scatter([i for i in range(1,6)],test_f1,c="red",label="test_f1")
plt.ylabel("f1")
plt.xlabel("fold")
plt.ylim(0,1)
plt.legend()
print("average train f1:",np.average(train_f1))
print("average test f1:",np.average(test_f1))
```

```python
plt.scatter([i for i in range(1,6)],train_acc,c="blue",label="train_acc")
plt.scatter([i for i in range(1,6)],test_acc,c="red",label="test_acc")
plt.ylabel("acc")
plt.xlabel("fold")
plt.ylim(0,1)
plt.legend()
print("average train accuracy:",np.average(train_acc))
print("average test accuracy:",np.average(test_acc))
```

Confusion matrix and other results of fold 0

```python
data_dir = r"C:\MTP\new_model\features\spectrogram"

class_names = sorted(os.listdir(data_dir))

class_to_idx = {name: i for i, name in enumerate(class_names)}

idx_to_class = {i: name for name, i in class_to_idx.items()}
```

```python
with open(r"C:\MTP\new_model\transformerencoder_model_spec\results_5folds\fold_CM\fold_0_confusion.npy","rb") as f:
    spec_cm=np.load(f)
```

```python
spec_cm
```

```python
plt.figure(figsize=(10,8))
cm_norm = spec_cm.astype("float") / spec_cm.sum(axis=1)[:, None]
sns.heatmap(
    cm_norm,
    cmap="Reds",
    xticklabels=class_names,
    yticklabels=class_names
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")

plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

cm = np.array(spec_cm)

TP = np.diag(cm)

FP = cm.sum(axis=0) - TP
FN = cm.sum(axis=1) - TP

precision = TP / (TP + FP + 1e-10)
recall    = TP / (TP + FN + 1e-10)

f1 = 2 * (precision * recall) / (precision + recall + 1e-10)
```

```python
plt.figure(figsize=(8,6))
idx = np.argsort(f1)

sorted_f1 = f1[idx]
sorted_names = np.array(class_names)[idx]

plt.figure(figsize=(12,8))

plt.barh(sorted_names, sorted_f1)

plt.xlabel("F1 Score")
plt.title("Per-Class F1 Score (sorted)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

num_classes = len(class_names)

# generate different colors
colors = plt.cm.tab20(np.linspace(0, 1, num_classes))

plt.figure(figsize=(12,8))

for i in range(num_classes):
    plt.scatter(
        recall[i],
        precision[i],
        color=colors[i],
        label=class_names[i],
        s=50
    )

plt.xlabel("Recall")
plt.ylabel("Precision")

plt.title("Precision vs Recall per Class")

plt.grid(True)

plt.xlim(0.2,1.1)
plt.ylim(0.4,1.1)

plt.legend(
    bbox_to_anchor=(1.05,1),
    loc="upper left",
    fontsize=8
)

plt.tight_layout()

plt.show()
```
